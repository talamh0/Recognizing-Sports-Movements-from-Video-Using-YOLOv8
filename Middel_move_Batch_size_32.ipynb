```python
from ultralytics import YOLO
```

```python
model1 = YOLO("yolov8n.pt")
model1.train(
    data= '/content/drive/MyDrive/Middle_move_dataset/data.yaml',
    epochs=100,  # Set a high number; early stopping will stop it earlier if needed
    batch=32,
    name="model_middel_frames",
    patience=10, # Stop if no improvement in validation loss for 10 epochs
    project="/content/drive/MyDrive/YOLO_training_Middlet_move_batch_32"
)
```

```python
# TESTING THE MODLE

# Load the trained models
model1 = YOLO("/content/drive/MyDrive/YOLO_training_Middlet_move_batch_32/model_middel_frames/weights/best.pt")
```

```python
# Run inference on test images
results1 = model1.predict(source='/content/drive/MyDrive/Middle_move_dataset/test/images', save=True)
```

```python
# Evaluate IOU for each model
data_yaml='/content/drive/MyDrive/Middle_move_dataset/data.yaml'
metrics1 = model1.val(data= data_yaml, split='test')

# Overall metrics
print(f"mAP@50: {metrics1.box.map50:.3f}")
print(f"mAP@50-95: {metrics1.box.map:.3f}")
print(f"Mean Precision: {metrics1.box.mp:.3f}")
print(f"Mean Recall: {metrics1.box.mr:.3f}")

# Per-class mAP
class_maps_3 = metrics1.box.maps
for i, ap in enumerate(class_maps_3):
    print(f"Class {i} ({model1.names[i]}): mAP = {ap:.2f}")
```
