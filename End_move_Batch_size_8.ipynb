```python
from ultralytics import YOLO
# Train Model 3 (Ending Frames)
model3 = YOLO("yolov8n.pt")
model3.train(
    data="/content/drive/MyDrive/End_move.v3i.yolov8/data.yaml",
    epochs=100,
    batch=8,
    name="model_ending_frames",
    patience=10,
    project="/content/drive/MyDrive/YOLO_Training_Results"
)
```

```python
from ultralytics import YOLO
# Load the trained model from Drive
model3 = YOLO("/content/drive/MyDrive/YOLO_Training_Results/model_ending_frames/weights/best.pt")

# Run inference on test images
results3 = model3.predict(
    source="/content/drive/MyDrive/End_move.v3i.yolov8/test/images",
    save=True
)
```

```python
# Evaluate model on Test set
metrics3 = model3.val(data="/content/drive/MyDrive/End_move.v3i.yolov8/data.yaml", split='test')

# Overall metrics
print(f"mAP@50: {metrics3.box.map50:.3f}")
print(f"mAP@50-95: {metrics3.box.map:.3f}")
print(f"Mean Precision: {metrics3.box.mp:.3f}")
print(f"Mean Recall: {metrics3.box.mr:.3f}")

# Per-class mAP
class_maps_3 = metrics3.box.maps
for i, ap in enumerate(class_maps_3):
    print(f"Class {i} ({model3.names[i]}): mAP = {ap:.2f}")
```
