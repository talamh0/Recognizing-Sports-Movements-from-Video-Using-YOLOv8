```python
# Load the trained models

model1 = YOLO("/content/drive/MyDrive/NEW YOLO_training_First_move_batch_32/Model_performance_metrics/weights/best.pt")
model2 = YOLO("/content/drive/MyDrive/YOLO_training_Middle_move_batch_32/Model_performance_metrics/weights/best.pt")

```

```python
model3 = YOLO("/content/drive/MyDrive/bestYOLO_training_End_move_batch_8/Model_performance_metrics/model_ending_frames/weights/best.pt")
```

```python
# Evaluate IOU for each model
print("Model 1 Start move:")
metrics1 = model1.val(data='/content/drive/MyDrive/First_move_dataset/data.yaml', split='test')
print("--------------------------------------------------------------------------------------------------------------------------------------------------------")

print("Model 2 Middle move:")
metrics2 = model2.val(data='/content/drive/MyDrive/Middle_move_dataset/data.yaml', split='test')
print("--------------------------------------------------------------------------------------------------------------------------------------------------------")

print("Model 3 End move:")
metrics3 = model3.val(data="/content/drive/MyDrive/End_move.v3i.yolov8/data.yaml", split='test')

```

```python
class_maps_1 = metrics1.box.maps
class_maps_2 = metrics2.box.maps
class_maps_3 = metrics3.box.maps # mAP for each class
```

```python
# Extract mAP results
map50_1 = metrics1.box.map50  # mAP@50
map50_2 = metrics2.box.map50
map50_3 = metrics3.box.map50

map5095_1 = metrics1.box.map  # mAP@50-95
map5095_2 = metrics2.box.map
map5095_3 = metrics3.box.map
```

```python
# Check model complexity (FLOPs)
flops1 = model1.info(verbose=True)
flops2 = model2.info(verbose=True)
flops3 = model3.info(verbose=True)
```

```python
# Overall comparison
import pandas as pd
import matplotlib.pyplot as plt

model_names = ['Start Frames', 'Middle Frames', 'End Frames']
map50 = [map50_1, map50_2, map50_3]
map5095 = [map5095_1, map5095_2, map5095_3]


df_overall = pd.DataFrame({
    'Model': model_names,
    'mAP@50': map50,
    'mAP@50-95': map5095
})
print("Overall Comparison Table:")
print(df_overall)

df_overall.set_index('Model').plot(kind='bar', figsize=(8, 5))
plt.title("Comparison of YOLOv8 Models")
plt.ylabel("Score")
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```
